```python
import matplotlib.pyplot as plt
import math as m

def gradient_descent(x, y, eta):

    function_val = []
    for _ in range(50):
        f = x ** 2 + 2 * (y ** 2) + 2 * m.sin(2 * m.pi * x) * m.sin(2 * m.pi * y)
        grad_x = 2*x + m.sin(2 * m.pi * y) * 4*m.pi*m.cos(2 * m.pi * x)
        grad_y = 4*y + 2*m.sin(2 * m.pi * x) * 2*m.pi*m.cos(2 * m.pi * y)

        x -= eta * grad_x
        y -= eta * grad_y
        function_val.append(f)
    
    print('x: {}\ny: {}'.format(x, y))

    print('f final: %f \n' % function_val[49])

    return function_val

def plot(result):
    
    plt.scatter(list(range(0, 50)), result, s=8)
    plt.plot(list(range(0, 50)), result)

    plt.xlabel('number of iterations')
    plt.ylabel('f')
    plt.show()


if __name__ == '__main__':

    x_0 = .1
    y_0 = .1
    eta1 = .01
    eta2 = .1

    result1 = gradient_descent(x_0, y_0, eta1)
    result2 = gradient_descent(x_0, y_0, eta2)

    plot(result1)
    plot(result2)
    
    gradient_descent(0.1, 0.1, 0.01)
    gradient_descent(1, 1, 0.01)
    gradient_descent(-0.5, -0.5, 0.01)
    gradient_descent(-1, -1, 0.01)
    print()
    gradient_descent(0.1, 0.1, 0.1)
    gradient_descent(1, 1, 0.1)
    gradient_descent(-0.5, -0.5, 0.1)
    gradient_descent(-1, -1, 0.1)
```
